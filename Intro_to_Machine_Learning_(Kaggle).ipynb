Seu primo ganhou milhões de dólares especulando no mercado imobiliário. Ele se ofereceu para se tornar seu parceiro de negócios devido ao seu interesse em ciência de dados. Ele fornecerá o dinheiro e você fornecerá modelos que prevêem quanto valem as várias casas.

Você pergunta ao seu primo como ele previu os valores dos imóveis no passado. e ele diz que é apenas intuição. Porém, mais questionamentos revelam que ele identificou padrões de preços de casas que viu no passado e usa esses padrões para fazer previsões para novas casas que está considerando.

O aprendizado de máquina funciona da mesma maneira. Começaremos com um modelo chamado Árvore de Decisão. Existem modelos mais sofisticados que fornecem previsões mais precisas. Mas as árvores de decisão são fáceis de entender e são o bloco de construção básico para alguns dos melhores modelos em ciência de dados.

Para simplificar, começaremos com a árvore de decisão mais simples possível.

image.png

Ele divide as casas em apenas duas categorias. O preço previsto para qualquer casa em consideração é o preço médio histórico de casas na mesma categoria.

Usamos os dados para decidir como dividir as casas em dois grupos e, em seguida, novamente para determinar o preço previsto em cada grupo. Esta etapa de captura de padrões de dados é chamada de ajuste ou treinamento do modelo. Os dados usados ​​para ajustar o modelo são chamados de dados de treinamento.

Os detalhes de como o modelo é ajustado (por exemplo, como dividir os dados) são complexos o suficiente para que possamos salvá-los para mais tarde. Depois de ajustar o modelo, você pode aplicá-lo a novos dados para prever os preços de casas adicionais.

Qual das duas árvores de decisão a seguir tem mais probabilidade de resultar do ajuste dos dados de treinamento imobiliário?

image.png

A árvore de decisão à esquerda (Árvore de decisão 1) provavelmente faz mais sentido, porque captura a realidade de que casas com mais quartos tendem a ser vendidas a preços mais altos do que casas com menos quartos. A maior deficiência desse modelo é que ele não captura a maioria dos fatores que afetam o preço da casa, como número de banheiros, tamanho do lote, localização, etc.

Você pode capturar mais fatores usando uma árvore que tenha mais "divisões". Estas são chamadas de árvores "mais profundas". Uma árvore de decisão que também considera o tamanho total do lote de cada casa pode ser assim:

image.png

Seu conjunto de dados tinha muitas variáveis ​​para resolver, ou até mesmo para visualizar bem. Como você pode reduzir essa quantidade esmagadora de dados a algo que você possa entender?

Começaremos escolhendo algumas variáveis ​​usando nossa intuição. Os cursos posteriores mostrarão técnicas estatísticas para priorizar variáveis ​​automaticamente.

Para escolher variáveis ​​/ colunas, precisaremos ver uma lista de todas as colunas no conjunto de dados. Isso é feito com a propriedade de colunas do DataFrame (a linha inferior do código abaixo).

```python
import pandas as pd
```

```python
# instanciando o conjunto de dados
melbourne_houses = pd.read_csv('melb_data.csv.zip', compression = 'zip')

melbourne_houses.columns # visualizando as colunas presentes
```

```python
''' Os dados de Melbourne têm alguns valores ausentes (algumas casas para as quais algumas variáveis ​​não foram registradas).
Aprenderemos a lidar com valores ausentes em um tutorial posterior.
Seus dados de Iowa não têm valores ausentes nas colunas que você usa.
Portanto, tomaremos a opção mais simples por enquanto, e retiraremos as casas de nossos dados.
Não se preocupe muito com isso agora, embora o código seja:'''

# a função ".dropna()" elimina os valores ausentes (pense em "na" como "não disponível")
melbourne_houses = melbourne_houses.dropna(axis=1)
```

```python
# visualizando se os dados ausentes/faltantes foram eliminados
melbourne_houses.isnull().sum()
```

Existem muitas maneiras de selecionar um subconjunto de seus dados. O curso [**Pandas**](https://www.kaggle.com/learn/pandas) cobre isso com mais profundidade, mas vamos nos concentrar em duas abordagens por enquanto.

1) Notação de pontos, que usamos para selecionar o "alvo de previsão"<br>
2) Selecionando com uma lista de colunas, que usamos para selecionar os "recursos"

<font size=4>***Selecionando o alvo de previsão***</font>

Você pode retirar uma variável com _notação de ponto_. Esta única coluna é armazenada em uma **série**, que é amplamente como um DataFrame com apenas uma única coluna de dados.

Usaremos a notação de ponto para selecionar a coluna que queremos prever, que é chamada de alvo de previsão. Por convenção, o alvo de previsão é chamado de `y`. Portanto, o código de que precisamos para salvar os preços das casas nos dados de Melbourne é

```python
# instanciando a variável alvo (target)
y = melbourne_houses['Price'] # podemos também instanciar dessa forma abaixo
# y = melbourne_houses.Price
```

<font size=4>***Escolhendo  os "Recursos (colunas)"***</font>

As colunas que são inseridas em nosso modelo (e posteriormente usadas para fazer previsões) são chamadas de "recursos". Em nosso caso, essas seriam as colunas usadas para determinar o preço da casa. Às vezes, você usará todas as colunas, exceto a coluna alvo (y). Outras vezes, você ficará melhor com menos recursos.

Por enquanto, vamos construir um modelo com apenas alguns recursos. Posteriormente, você verá como iterar e comparar modelos construídos com diferentes recursos.

Selecionamos vários recursos fornecendo uma lista de nomes de colunas entre colchetes. Cada item dessa lista deve ser uma string (com aspas).

Aqui está um exemplo:

```python
# selecionando as features de interesse
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

> Por convenção, esses dados são chamados de `X`.

```python
# instanciando as colunas (recursos) que explica a nossa variável alvo
X = melbourne_houses[melbourne_features]
```

Vamos revisar rapidamente os dados que usaremos para prever os preços das casas usando o método `describe` e o método `head`, que mostra as poucas linhas superiores.

```python
X.describe()
```

```python
X.head() # por padrão, retorna as 5 primeiras linhas
```

Verificar visualmente seus dados com esses comandos é uma parte importante do trabalho de um cientista de dados. Freqüentemente, você encontrará surpresas no conjunto de dados que merecem uma inspeção mais detalhada.

<font size=4>***Construindo Seu Modelo***</font>

Você usará a biblioteca `scikit-learn` para criar seus modelos. Ao codificar, esta biblioteca é escrita como sklearn, como você verá no código de amostra.

_Scikit-learn é facilmente a biblioteca mais popular para modelar os tipos de dados normalmente armazenados em DataFrames._

As etapas para construir e usar um modelo são:

- **Definir**: que tipo de modelo será? Uma árvore de decisão? Algum outro tipo de modelo? Alguns outros parâmetros do tipo de modelo também são especificados.
- **Ajustar**: captura os padrões dos dados fornecidos. Este é o cerne da modelagem.
- **Prever**: exatamente o que parece
- **Avaliar**: determine o quão precisas são as previsões do modelo.

Aqui está um exemplo de definição de um modelo de árvore de decisão com scikit-learn e ajustá-lo com os recursos e a variável de destino.

```python
# importantando a lib para o nosso modelo
from sklearn.tree import DecisionTreeRegressor # árvore para regressão
```

```python
# instanciando o modelo
melbourne_model = DecisionTreeRegressor(random_state=1)
# obs: especifique um número para random_state para garantir os mesmos resultados a cada execução

# treinando o modelo 
melbourne_model.fit(X, y)
```

Muitos modelos de aprendizado de máquina permitem alguma aleatoriedade no treinamento do modelo. Especificar um número para random_state garante que você obtenha os mesmos resultados em cada execução. Isso é considerado uma boa prática. Você pode usar qualquer número, e a qualidade do modelo não dependerá significativamente do valor que você escolher.

Agora temos um modelo ajustado que podemos usar para fazer previsões.

Na prática, você desejará fazer previsões para novas casas chegando ao mercado, em vez das casas para as quais já temos preços. Mas faremos previsões para as primeiras linhas dos dados de treinamento para ver como a função de previsão funciona.

```python
print('Fazendo previsões para as 5 casas a seguir:')
print(X.head())
print('\nAs predições dos preços são:')
print(melbourne_model.predict(X.head()))
print('\nPreço real:')
print(melbourne_houses['Price'].head())
```

## <font color='orange'>**04 - Validação do modelo**</font>

Você construiu um modelo. Mas isso é bom?

Nesta lição, você aprenderá a usar a validação de modelo para medir a qualidade de seu modelo. Medir a qualidade do modelo é a chave para melhorar iterativamente seus modelos.

<font size=4>***O que é validação de modelo***</font>

Você vai querer avaliar quase todos os modelos que construiu. Na maioria dos aplicativos (embora não em todos), a medida relevante da qualidade do modelo é a precisão preditiva. Em outras palavras, as previsões do modelo serão próximas do que realmente acontece.

Muitas pessoas cometem um grande erro ao medir a precisão preditiva. Eles fazem previsões com seus dados de treinamento e comparam essas previsões aos valores de destino nos dados de treinamento. Você verá o problema com essa abordagem e como resolvê-lo em um momento, mas vamos pensar sobre como faríamos isso primeiro.

Primeiro, você precisa resumir a qualidade do modelo de uma forma compreensível. Se você comparar os valores residenciais previstos e reais para 10.000 casas, provavelmente encontrará uma mistura de previsões boas e ruins. Olhar em uma lista de 10.000 valores previstos e reais seria inútil. Precisamos resumir isso em uma única métrica.

Existem muitas métricas para resumir a qualidade do modelo, mas começaremos com uma chamada `Erro Médio Absoluto` (também chamado de **MAE**). Vamos decompor essa métrica começando com a última palavra, erro.

O erro de previsão para cada casa é:

`erro = real-previsto`

Portanto, se uma casa custar `$ 150.000` e você previu que custaria `$ 100.000`, o erro será `$ 50.000`.

Com a métrica MAE, pegamos o valor absoluto de cada erro. Isso converte cada erro em um número positivo. Em seguida, calculamos a média desses erros absolutos. Esta é a nossa medida de qualidade do modelo. Em português simples, pode ser dito como:

    Em média, nossas previsões estão erradas em cerca de X.

Para calcular o MAE, primeiro precisamos de um modelo.

```python
# importando a lb para modelagem de dados
import pandas as pd

# instanciando o conjunto de dados
melbourne_houses = pd.read_csv('melb_data.csv.zip', compression = 'zip')

melbourne_houses.columns # visualizando as colunas presentes
```

```python
# selecionando as features de interesse
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
# instanciando as colunas (recursos) que explica a nossa variável alvo
X = melbourne_houses[melbourne_features]

# instanciando a variável alvo (target)
y = melbourne_houses['Price']
```

```python
# importando a lib
from sklearn.tree import DecisionTreeRegressor 

# definindo o modelo
melbourne_model = DecisionTreeRegressor()

# treinando o modelo
melbourne_model.fit(X, y)
```

Assim que tivermos um modelo, veja como calculamos o erro médio absoluto:

```python
# importando a função para calcular o erro médio absoluto
from sklearn.metrics import mean_absolute_error

# calculando o o erro médio absoluto do modelo
predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

<font size=4>***O problema com pontuações "n-amostra"***</font>

A medida que acabamos de calcular pode ser chamada de pontuação "dentro da amostra". Usamos uma única "amostra" de casas tanto para construir o modelo quanto para avaliá-lo. Veja por que isso é ruim.

Imagine que, no grande mercado imobiliário, a cor da porta não esteja relacionada ao preço da casa.

No entanto, na amostra de dados que você usou para construir o modelo, todas as casas com portas verdes eram muito caras. O trabalho do modelo é encontrar padrões que prevejam os preços das casas, para que ele veja esse padrão e sempre preveja preços altos para casas com portas verdes.

Como esse padrão foi derivado dos dados de treinamento, o modelo parecerá preciso nos dados de treinamento.

Mas se esse padrão não se mantiver quando o modelo vê novos dados, o modelo seria muito impreciso quando usado na prática.

Como o valor prático dos modelos vem de fazer previsões sobre novos dados, medimos o desempenho em dados que não foram usados ​​para construir o modelo. A maneira mais direta de fazer isso é excluir alguns dados do processo de construção do modelo e, em seguida, usá-los para testar a precisão do modelo em dados que ele nunca viu antes. Esses dados são chamados de **dados de validação**.

<font size=4>***Codificando***</font>

A biblioteca scikit-learn tem uma função `train_test_split` para dividir os dados em duas partes. Usaremos alguns desses dados como dados de treinamento para ajustar o modelo e usaremos os outros dados como dados de validação para calcular `mean_absolute_error`.

Aqui está o código:

```python
# importando a função para dividir o conjunto de dados
from sklearn.model_selection import train_test_split
```

```python
# dividir os dados em dados de treinamento e validação, para recursos e destino
# A divisão é baseada em um gerador de números aleatórios. Fornecendo um valor numérico para
# o argumento random_state garante que obteremos a mesma divisão toda vez que execute o script
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

## <font color='orange'>**05 - Underfitting e Overfitting**</font>

Ao final desta etapa, você entenderá os conceitos de `subajuste` (underfitting) e `sobreajuste` (overfitting) e poderá aplicar essas ideias para tornar seus modelos mais precisos.

<font size=4>***Experimentando com diferentes modelos***</font>

Agora que você tem uma maneira confiável de medir a precisão do modelo, pode experimentar modelos alternativos e ver qual oferece as melhores previsões. Mas que alternativas você tem para modelos?

Você pode ver na [documentação](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html) do scikit-learn que o modelo de árvore de decisão tem muitas opções (mais do que você deseja ou precisa por muito tempo). As opções mais importantes determinam a profundidade da árvore. Lembre-se da [primeira lição](https://www.kaggle.com/dansbecker/how-models-work) deste curso que a profundidade de uma árvore é uma medida de quantas divisões ela faz antes de chegar a uma previsão. Esta é uma árvore relativamente rasa

image.png

Na prática, não é incomum que uma árvore tenha 10 divisões entre o nível superior (todas as casas) e uma folha. Conforme a árvore fica mais profunda, o conjunto de dados é dividido em folhas com menos casas. Se uma árvore tiver apenas 1 divisão, ela divide os dados em 2 grupos. Se cada grupo for dividido novamente, obteremos 4 grupos de casas. Dividir cada um deles novamente criaria 8 grupos. Se continuarmos dobrando o número de grupos adicionando mais divisões em cada nível, teremos 210 grupos de casas no momento em que chegarmos ao 10º nível. Isso é 1024 folhas.

Quando dividimos as casas entre muitas folhas, também temos menos casas em cada folha. Folhas com muito poucas casas farão previsões muito próximas dos valores reais dessas casas, mas podem fazer previsões pouco confiáveis ​​para novos dados (porque cada previsão é baseada em apenas algumas casas).

Este é um fenômeno denominado `overfitting`, em que um modelo corresponde aos dados de treinamento quase perfeitamente, mas se sai mal na validação e em outros dados novos. Por outro lado, se deixarmos nossa árvore muito rasa, isso não dividirá as casas em grupos muito distintos.

No extremo, se uma árvore divide as casas em apenas 2 ou 4, cada grupo ainda tem uma grande variedade de casas. As previsões resultantes podem estar distantes para a maioria das casas, até mesmo nos dados de treinamento (e também será ruim na validação pelo mesmo motivo). Quando um modelo falha em capturar distinções e padrões importantes nos dados, ele tem um desempenho ruim mesmo em dados de treinamento, isso é chamado de **subajuste** (underfitting).

Como nos preocupamos com a precisão dos novos dados, que estimamos a partir de nossos dados de validação, queremos encontrar o ponto ideal entre `underfitting` e `overfitting`. 

Visualmente, queremos o ponto baixo da curva de validação (vermelha) na figura abaixo.

image.png

<font size=4>***Exemplo***</font>

Existem algumas alternativas para controlar a profundidade da árvore e muitas permitem que algumas rotas através da árvore tenham maior profundidade do que outras rotas. 

Mas o argumento `max_leaf_nodes` fornece uma maneira muito sensata de controlar overfitting vs underfitting. Quanto mais folhas permitimos que o modelo faça, mais nos movemos da área de underfitting no gráfico acima para a área de overfitting.

Podemos usar uma função de utilidade para ajudar a comparar as pontuações MAE de diferentes valores para **max_leaf_nodes**:

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, X_train, X_test, y_train, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(X_train, y_train)
    preds_val = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds_val)
    return(mae)
```

> Os dados são carregados em X_train, X_test, y_train e y_test usando o código que você já viu (e que já escreveu).

Podemos usar um loop for para comparar a precisão dos modelos construídos com diferentes valores para `max_leaf_nodes`.

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, X_train, X_test, y_train, y_test)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

> Das opções listadas, 500 é o número ideal de folhas.

<font size=5>***Conclusão***<font>

Aqui está a conclusão, os modelos podem sofrer de:

- `Overfitting`: captura de padrões espúrios que não ocorrerão no futuro, levando a previsões menos precisas ou
- `Underfitting`: falha em capturar padrões relevantes, novamente levando a previsões menos precisas.

Usamos dados de validação, que não são usados ​​no treinamento do modelo, para medir a precisão de um modelo candidato. Isso nos permite experimentar vários modelos candidatos e manter o melhor.
